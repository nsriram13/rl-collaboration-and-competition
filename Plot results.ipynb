```python
import os
import pickle

import numpy as np
import matplotlib.pyplot as plt
from deeprl_cc.utils import smooth

plt.style.use('ggplot')
```

```python
if os.path.exists('./checkpoints/scores.pkl'):
    with open('./checkpoints/scores.pkl', 'rb') as f:
        scores = pickle.load(f)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(range(len(scores)), scores, color='#60BD68')
plt.plot(range(len(scores)), smooth(scores, radius=10), color='#B276B2')
plt.xlabel('Number of episodes trained')
plt.ylabel('Score')
```
